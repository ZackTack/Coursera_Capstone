```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd

url = "http://www.qianmu.org/2019USNEWS%E7%BE%8E%E5%9B%BD%E8%AE%A1%E7%AE%97%E6%9C%BA%E7%A7%91%E5%AD%A6%E6%8E%92%E5%90%8D"
result = requests.get(url)
soup = BeautifulSoup(result.content,'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table))[0]
df
```

```python
df.rename(columns=df.iloc[0],inplace=True)
df.drop(0,inplace=True)
```

```python
df.drop(df.columns[1],axis=1,inplace=True)
```

```python
df.columns=['Rank','University','Rate','City/State']
df
```

```python
result_df = pd.DataFrame()
for i in range(1,10):
    search_url = f"https://yocket.in/university-application-deadlines?university_id=0&course_group_id=4&level=masters&deadline_sub_type_id%5B%5D=0&deadline_sub_type_id%5B%5D=3&deadline_sub_type_id%5B%5D=4&deadline_sub_type_id%5B%5D=7&deadline_sub_type_id%5B%5D=22&deadline_sub_type_id%5B%5D=23&deadline_sub_type_id%5B%5D=29&deadline_sub_type_id%5B%5D=31&deadline_sub_type_id%5B%5D=32&deadline_sub_type_id%5B%5D=34&deadline_sub_type_id%5B%5D=46&deadline_sub_type_id%5B%5D=47&deadline_sub_type_id%5B%5D=48&page={i}"
    result = requests.get(search_url)
    soup = BeautifulSoup(result.content,'lxml')
    table = soup.find_all('table')[0]
    df2 = pd.read_html(str(table))[0]
    df2 = df2.merge(df,on='University')
    result_df = pd.concat([result_df,df2])
    result_df = result_df[result_df.Rank != 'RNP']
```
